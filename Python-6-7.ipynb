### CSV

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3,\n'
        'x,y,z,\n'
        'a,b,c,\n'
        'c,d,3,')
#\n depicts next line
#in csv file our data is usually in this format
```

```python
type(data)
```

```python
# StringIO()  Text I/O implementation using an in-memory buffer
#basically isko jo bhi data denge oski memory buffer bana dega

pd.read_csv(StringIO(data))
#and then after memory buffer is created then read this using read_csv
#it will understand that col1 , col2 and col3 are index and rest are values
```

```python
#Learning some parameters in read_csv
#use col says from all the cols in csv file u can get some specifies cols

df1=pd.read_csv(StringIO(data),usecols=['col1','col3'])
```

```python
df1
```

```python
#to convert whole data into a csv file then use the function given below
df1.to_csv('Python-6.csv')
#write the file name that u wanna give to the file
```

```python
#specifing the data type of col
#basically we will provide the data type

data=('a,b,c,d,\n'
      '1,2,3,4,\n'
      '5,6,7,8,\n'
      '9,10,11,')
```

```python
print(data)
```

```python
df2=pd.read_csv(StringIO(data),dtype=object)
#give memory buffer to data and make data type as object
#object means all values will be strings
```

```python
df2
```

```python
df2['a'][1]
#verifying all numbers are strings
```

```python
#Now specifying each col of diff data type
df3=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64','d':object})
#int is dynamically allocating memory whereas int64 has fixed 64 bits or 8 bytes of memory allocation
```

```python
df3
```

```python
type(df3['d'][1])
```

```python
#checking all the data types of col
df3.dtypes
```

```python
data1=('index,a,b,c,\n'
       '4,apple,bat,5.7,\n'
       '8,orange,cow,10,')
```

```python
#pd.read_csv(StringIO(data1))
#basically i wanted that 4 and 8 to be index col . By default the indexs are 0 and 1
#for that we use index_col=0
pd.read_csv(StringIO(data1),index_col=0)
# in here we are saying that take the first col as my index
```

```python
data2=('a,b,c,\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data2))
```

```python
#sometimes if we give numbers then it by default takes those numbers(here 4 and 8) as index 
# 4 and 8 must come under a so in order to resolve that issue we do
pd.read_csv(StringIO(data2),index_col=False)
```

```python
#combining usecols and index_cols
data3=('a,b,c,\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data3),usecols=['b','c'],index_col=False)
```

```python
#Quoting and Escape Characters. Very useful in NLP

data4 = 'a,b\n"hello, \\"bob\\", nice to see you",5'
#a,b and then new line
#hello and then escape characters
```

```python
pd.read_csv(StringIO(data4),escapechar='\\')
#here in escapechar just type the charcter u want to skip/escape
```

### Read Json to CSV

```python
#JSON is just like key value pairs
Data = '{"employee_name":"James","email":"james@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Sr. Developer"}]}'
df1=pd.read_json(Data)
#reading json data and converting into dataframe
#if u have nested info like job profile then i will not exactly get converted into cols
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
#this url is in json format and we have converted it into dataframe
#header=none will display col1 as 0 and col2 as 1 and so on
```

```python
df.head()
#we are reading data without downloading it
#structures comes only when u have key value pair in data
#if u have nested values then mushkil ho jayega to show it 
```

```python
#now we will convert the json into csv file
df.to_json("wine.csv")
#go and check the loc and u will find wine.csv file there
```

```python
df1
```

```python
#convert to JSON format
df1.to_json()
#first it goes to 1st col and creates key value pair of 0 and james
#simillarly for email
df1.to_json(orient='records')
#isse it will make record by record json file
```

```python
#convert json to diff json formats
df.to_json(orient="records")
#we got 0 : 1 .. 1: 14.23 ... 2:1.71 and for each record u will get in this way
```
