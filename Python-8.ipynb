## MatplotLib Tutorial
Matplotlib is a plotting library for Python programming language and its numerical mathematics extension Numpy. It provides on object-oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter,wxPython,QT, or GTK+

Some of the major Pros of Matplotib are:
- Generally easy to get started for simple plots
- Support for custom tables and texts
- Great control of every element in a fig
- High quality output in many formats
- Very customizable in general

### Scatter Plot

```python
#plotting using matplotlib

#plt scatter
#we are making scatter plot and it is used to scatter the values of x and y
plt.scatter(x,y,c='g')
#here c means colour and g means green
#s is the pixel value and it will be in numbers
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Graph in 2D')
#we need not write plt.show() to see the diagrams since we wrote taht %modulas... wali line in start
#u can right click the image and also save it 
#apart from that u can also do the following
plt.savefig('Test.png')
#this will save the fig with name test in png format
```

### Plot 

```python
##plt plot

plt.plot(x,y1,'r*--',linewidth=2,markersize=12)
#in plot we will get a straight line
#press shift tab to  know about the parameters and colours that u can choose from and their letters
#here if we dont write c='r' and directly write r then the line becomes red colour
# we want the line to be double dashed then we write 'r--'
# or u can also write linestyle='dashed'
# similarly 'r*--' is the line style we want
#u can also use 'ro' or 'ro--' .. basically o depicts the points in between
#markers size refers to size of points
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Graph')
```

### Creating subplots

```python
#subplot means within 1 diagram i can create multiple subplots within that
plt.subplot(2,2,1)
#3 basic parameters are required i.e number of rows and number of cols and last is position
#position basically says that in 1st postion i need my 1 diagram

plt.plot(x,y,'r--')
#then we went to 2nd index and plotted my 2nd graph and so on for others
plt.subplot(2,2,2)
plt.plot(x,y1,'g*-')
plt.subplot(2,2,3)
plt.plot(y1,x,'b')
plt.subplot(2,2,4)
plt.plot(x,x,'yo')
#scatter plot made
```

```python
#Computer the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
#0 to 4*np.pi and at stepsize of 0.1
y=np.sin(x)
#take all values of x and find value of sin of x and put them in y
plt.title("sine wave form")

#Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#inbuilt functions in numpy we have np.sin and np.cos
#subplots
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up the subplot grid that has height of 2 and width of 1
#and set the first such subplot as active
plt.subplot(2,1,1)

#make first plot
plt.plot(x,y_sin,'g--')
plt.title("Sine")

#set second plot as active and make second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r-')
plt.title("Cosine")
```

### Boxplots

```python
#selecting normal distribution in between 0 to std and std ranges between 1 to 4
#100 is step size 
#basically means it will loop between 0 to 1 and then 0 to 2 and so on
data=[np.random.normal(0,std,100) for std in range(1,4)]
#also see what does this data look like
#It creates a list of 3 arrays each containing 100 values sampled from normal distribution with mean=0 and std =1,2,3 

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
#boxplot helps you find percentiles
#circles are outliners
#sabse niche wali line is 0 percentile
#then 25 percentile 50 percentile 75 percentile and 100 percentile
#if u make patch_artist to be false then colour goes away
```

### Pie Chart

```python
#Data to plot
labels= 'Python','C++','Ruby','Java'
sizes=[215,130,245,210]
#basically size/total of all  sizes into hundred and it will be converted into percentage
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st slice..means how much far 1st slice will go
#if u give 0.4 then it will go more far
#u can give each slice some explode value as well
#for colors check list of colors available for pie and only that u can assign

#Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
#autoct means whatever slice u are making then in what format u want that
# 1.1f means all the percentage will be shown in floating format
# % means start of format specifier
# 1.1f means format with 1 digit before and 1 digit after the deicmal
# %% outputs % sign ...basically %% tells python to literally print % sign

plt.axis('equal')
#By default, matplotlib automatically scales axes to fit the data, which can distort shapes like circles aand sqs
#the above line preserves the actual aspect ratio
```
